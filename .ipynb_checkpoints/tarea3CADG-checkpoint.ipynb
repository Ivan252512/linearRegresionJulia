# Ejercicio 3

## $y = a b ^ x$ 
### $log(y) = log(a b ^ x) = log(a) + log (b ^ x) = log (y) = log(a) + x log (b )$

### Si ahora $y_1 = log(y)$, $b_1 = log(a)$ , $ x = x $ y $ m = log(b) $
### Llegamos a una ecuación de la forma $y_1 = mx + b_1$
### Por lo tanto $e^{b_1} = e^{log(a)}$ entonces $a = e ^ {b_1}$
### Y $e^{m} = e^{log(b)}$ entonces $b = e ^ {m}$

```julia
#Es lo mismo que uno, solo modificamos y en ls. renglones 13-15 y usamos que

# Función que leé los argumentos pasados en la terminal
function archivos(ARGS)
    @show ARGS
end

argumentos = archivos(ARGS)


#Ejercicio 2, se necesitan dos arreglos, "x" y "y".
function ls2(x,y)
    #Transoformación
    ylog = zeros(length(y))
    for i = 1:length(y)
        ylog[i] = log(y[i])
    end 
    #Regresamos a y b
    return m(x,ylog), b(x,ylog)
end

#Media
function media(x)
    xm = 0
    for i=1:length(x)
        xm += x[i]
    end
    return xm/length(x)
end

#Pendiente
function m(x,y)
    if length(x) == length(y)
        #Calculamos x * y
        xy = zeros(length(x))
        for i = 1:length(xy)
            xy[i] = x[i] * y[i]
        end

        #Calculamos xprom y yprom
        xprom = media(x)
        yprom = media(y)

        #Calculamos x ^ 2
        x2 = zeros(length(x))
        for i = 1:length(x)
            x2[i] = x[i] ^ 2
        end

        #Calculamos x ^ 2 prom
        x2prom = media(x2)

        #N
        n = length(x)

        #Regresamos m
        return (sum(xy) - n*xprom*yprom)/(sum(x2)-n*x2prom)
    end
    return
end

#b
function b(x,y)
    if length(x) == length(y)
        #yprom
        yprom = media(y)

        #Calculamos x ^ 2
        x2 = zeros(length(x))
        for i = 1:length(x)
            x2[i] = x[i] ^ 2
        end

        #xprom
        xprom = media(x)

        #Calculamos x * y
        xy = zeros(length(x))
        for i = 1:length(xy)
            xy[i] = x[i] * y[i]
        end

        #x ^ 2 prom
        x2prom = media(x2)

        #N
        n = length(x)

        return (yprom*sum(x2) - xprom*sum(xy)) / (sum(x2) - n*x2prom)
    end
end
```

```julia
# Solo se modifica una función
#Ejercicio 3, se necesitan dos arreglos, "x" y "y".
#a y b son las constantes para la translación de datos.
function ls2(x,y,a,b) 
    #Linealización
    if length(x)==length(y)
        xtrans = zeros(length(x))
        ytrans = zeros(length(x))
        ylog = zeros(length(y))
        for i = 1:length(y)
            #Translación
            xtrans[i] = x[i] + a
            ytrans[i] = y[i] + b
            #Linealización   
            ylog[i] = log(ytrans[i])
        end 
        return m(x,ylog), b(x,ylog)
    end
    return
end
```
